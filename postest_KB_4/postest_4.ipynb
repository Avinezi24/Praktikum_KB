```python
# Import Library

import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("cars.csv")
df.head(10)
```

```python
# Mengambil beberapa kolom yang diperlukan

df_baru = pd.DataFrame(df.loc[:,"manufacturer_name":"up_counter"])
df_baru.head()
```

### Data Split Proporsi 70:30

```python
from sklearn.model_selection import train_test_split
# Subset feature dan target
x = df_baru.iloc[:,:-1]
y = df_baru.iloc[:,-1]

x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.3)

print("Dimensi x_train: ", x_train.shape)
print("Dimensi y_train: ", y_train.shape)
print("Dimensi x_test: ", x_test.shape)
print("Dimensi y_test: ", y_test.shape)
```

### Cleaning Data NULL

```python
df_baru.isna().sum() # Melihat jumlah nilai null di dalam kolom
```

```python
from sklearn.impute import SimpleImputer 
cleanNull = SimpleImputer(strategy="mean") # Mengahapus nilai null menggunakan strategi mean
df_baru["engine_capacity"] = cleanNull.fit_transform(df_baru[["engine_capacity"]])
```

```python
df_baru.isna().sum() # Mengecek nilai null
```

### Cleaning Data Duplikat

```python
df_baru.duplicated().sum() # Mengecek nilai duplikat dalam dataset
```

```python
df_baru.drop_duplicates(inplace=True)
df_baru.duplicated().sum() # Melihat kembali jumlah nilai duplikat dengan menggunakan fungsi drop_duplicates()
```

### Normalisasi

```python
from sklearn.preprocessing import MinMaxScaler 
dfnormalisasi= df_baru[["odometer_value"]]
minMax= MinMaxScaler()
skalaNormalis = minMax.fit_transform(
    dfnormalisasi[["odometer_value"]]) 
dfnormalisasi_baru = pd.DataFrame(skalaNormalis,columns=["Setelah Normalisasi"])
dfnormalisasi = dfnormalisasi.join(dfnormalisasi_baru)
dfnormalisasi.head()
```

### Standarisasi

```python
from sklearn.preprocessing import StandardScaler
dfstandar=df_baru.select_dtypes(include='number')
```

```python
np.std(dfstandar) # Sebelum standarisasi
```

```python
ss = StandardScaler()
scale_standar = ss.fit_transform(dfstandar)
dfstandar_baru= pd.DataFrame(scale_standar, columns=["odometer_value_std","year_produced_std","engine_capacity_std", "price_usd_std","number_of_photos_std","up_counter_std"])
dfstandar = dfstandar.join(dfstandar_baru)
```

```python
np.std(dfstandar_baru) # Sesudah standarisasi
```

### Ganti tipedata salah satu atribut angka

```python
df_baru["up_counter"].dtypes # Tipe data sebelum diganti
```

```python
df_baru[['up_counter']] = df_baru[['up_counter']].astype("object") # Proses ganti type data
```

```python
df_baru.info() # Mengecek tipe data sebelum
```

### One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
OneHotEn = OneHotEncoder(sparse=False)
oneHot = OneHotEn.fit_transform(df_baru[["engine_type"]])
typemesin = df_baru.dropna(subset=['engine_type']).sort_values(by=['engine_type'])['engine_type'].unique() # Mengambil data kategori
df_OneHotencoding= pd.DataFrame(oneHot, columns=typemesin)
df_OneHotencoding.head()
```

```python
df_baru = df_baru.join(df_OneHotencoding) 
df_baru.head()
```
